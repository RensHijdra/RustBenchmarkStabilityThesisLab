```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iter_samp_combinations = [(range(5), [300] * 5, '\\\\', '#d95f02'), (range(8), [150] * 8, 'oo', '#1b9e77'), (range(15), [100] * 15, '//', '#7570b3')]
ax = plt.subplot(111)
for x, y, hatch, color in iter_samp_combinations:
    ax.barh(x, y, fill=False, hatch=hatch, edgecolor=color)
plt.xlabel("Number of samples")
plt.ylabel("Number of iterations")
plt.yticks(range(15), np.arange(0, 15) + 1)
plt.legend(["(300, 5)", "(150, 8)", "(100, 15)"])
plt.savefig('explanation_samples_iters.png', dpi=600)
```

```python
import random
colors = ['#e41a1c',
'#377eb8',
'#4daf4a',
'#984ea3',
'#ff7f00']


ax = plt.subplot(111)
for i in range(3):
        random.shuffle(colors)
        for b in range(5):
            ax.barh(i, 300, left=b * 300 + i * 1500, color=colors[b])
ax.set_xlabel("Total samples taken over time")
ax.set_ylabel("Iteration")
ax.set_yticks(range(0, 3), map(str, range(1,4)))
ax.grid(which='major')
plt.savefig('iter_over_time_example.png', dpi=600)
```
